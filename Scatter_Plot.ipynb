#**Scatter Plot**

```python
import matplotlib.pyplot as plt
study_hours = [1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0]
exam_score = [50,55,60,50,75,80,65,80,70,60,75,90]
colours = [42,52,62,74,10,65,65,74,30,40,61,53]
plt.scatter(study_hours,exam_score,marker="x",s=100,c=colours,cmap="viridis")
plt.title("Study Hours vs. Exam Scores",fontsize=15)
plt.xlabel("Study Hours",fontsize=12)
plt.ylabel("Exam Score",fontsize=12)
t=plt.colorbar()
t.set_label("Color Bar")
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
study_hours = [1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0]
exam_score = [50,55,60,50,75,80,65,80,70,60,75,90]
plt.scatter(study_hours,exam_score)
plt.plot(study_hours,exam_score,alpha=0.6)
plt.bar(study_hours,exam_score,width=0.3,color="skyblue",alpha=0.5)
plt.title("Study Hours vs. Exam Scores",fontsize=15)
plt.xlabel("Study Hours",fontsize=12)
plt.ylabel("Exam Score",fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
