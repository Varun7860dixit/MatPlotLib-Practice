#**Line Plot**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,5),facecolor="lightgrey")

months = ["January","February","March","April","May","June","July","August","September","October","November","December"]
temp = [5,7,12,18,24,29,32,31,27,20,13,7]

plt.plot(months,temp,color="blue",linestyle="--",marker="o",linewidth=2,markersize=10,markeredgecolor="blue",markeredgewidth=7,alpha=0.3,label="Temperature (°C)")

plt.title("Monthly Average Temperature (°C)",fontsize=15)
plt.xlabel("Month",fontsize=12)
plt.ylabel("Temperature (°C)",fontsize=12)

plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.plot(x,y,linestyle="--",linewidth=2,label="Sinx",color="green")

plt.title("Graph of Sinx",fontsize=15)
plt.xlabel("Value of x",fontsize=12)
plt.ylabel("Value of y",fontsize=12)

plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
