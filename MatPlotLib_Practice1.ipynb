#          **Bar Plot**

####1. Population Growth Rate by Country (2022 Estimates)

```python
import matplotlib.pyplot as plt
country = ["India","China","USA","Indonesia","Pakistan"]
growth_rate = [0.95,0.03,0.59,0.81,2.00]
plt.bar(country,growth_rate,width=0.5,edgecolor="black",alpha=0.6,label="Population Growth Rate (%)")
plt.title(" Population Growth Rate by Country (2022 Estimates)",fontsize=14)
plt.xlabel("Country Name",fontsize=12)
plt.ylabel("Growth Rate (%)",fontsize=12)
plt.legend()
plt.show()
```

####2. Monthly Sales (USD)

```python
import matplotlib.pyplot as plt
import numpy as np
region = ["North","South","East","West"]
sales1 = [12500,9800,15600,11800]
sales2 = [10300,11200,13400,14500]
width = 0.3
p = np.arange(len(region))
s = [j+width for j in p]
plt.bar(p,sales1,color="b",label="Product A",width=0.3,edgecolor="black",alpha=0.5)
plt.bar(s,sales2,color="skyblue",label="Product B",width=0.3,edgecolor="black",alpha=0.6)
plt.title("Monthly Sales (USD)",fontsize=16)
plt.ylabel("Product A and B sales",fontsize=12)
plt.xlabel("Region",fontsize=12)
plt.xticks(p+width/2,region,rotation=25,fontsize=10)
plt.legend()
plt.show()
```
